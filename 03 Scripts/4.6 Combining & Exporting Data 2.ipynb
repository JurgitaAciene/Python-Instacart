# 03 Merging dfs prods and orders_products_combined

Q5 both dfs have same column 'product_id', which is probably th ebest way to combine. The number of rows differ significantly, but if we only need additional data from orders_products_combined to add to prods, that should be it. 

```python
# drop '_merge' column from df_orders_products_combined
df_orders_products_combined_2 = df_orders_products_combined.drop(columns = ['_merge'])
```

```python
# Testing merge w/o overwriting
pd.merge(df_prods,df_orders_products_combined_2,on = 'product_id', indicator = True)
```

```python
# Merging dfs prods and rders_products_combined_2
df_ords_prods_merged = df_prods.merge(df_orders_products_combined_2,on = 'product_id', indicator = True)
```
