# 02 Creating new variables with user defined If-Statement

```python
# Create a subset of 1,000,000
df=ords_prods_merge[:1000000]
```

```python
# Check shape
df.shape
```

```python
# Define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check counts
df['price_range'].value_counts(dropna = False)
```

```python
# Check most expensive
df['prices'].max()
```

# 04 Creating new variables with For-Loops

```python
# Cheking counts to determine busiest days
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create labels and new column 'busiest_day' with For-Loops Q1
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create new column
ords_prods_merge['busiest_day'] = result
```

```python
# Check counts
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

# 05 Create 'busiest_days' column and new labels Q2

```python
# Create new labels where 2 busiest days are "Busiest days" and 2 least busy are "Slowest days"
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Busiest days")
  elif value == 3 or value == 4:
    result.append("Slowest days")
  else:
    result.append("Regularly busy")
```

```python
# Create a new column
ords_prods_merge['busiest_days'] = result
```

```python
# Check counts Q3
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

# 06 Create 'busiest_period_of_day' column and new labels Q4

```python
# Cheking counts to determine busiest hours
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create new labels where 8 busiest hours are "Most orders", 8 least busy are "Fewest orders" and the remaining 8 are "Average orders".
result = []

for x in ords_prods_merge["order_hour_of_day"]:
  if x == 10 or x == 11 or x == 14 or x == 15 or x == 13 or x == 12 or x == 16 or x == 9:
    result.append("Most orders")
  elif x == 3 or x == 4 or x == 2 or x == 5 or x == 1 or x == 0 or x == 6 or x == 23:
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

```python
# Create a new column
ords_prods_merge['busiest_period_of_day'] = result
```

```python
# Check counts Q5
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
