# 02 Creating charts from exercise Q2

### Bar

```python
# Create Bar Chart
bar_orders_dow = ords_prods_custmrs['orders_day_of_week'].value_counts().sort_index().plot.bar(color = 'green')
```

### Scatterplot

```python
# Create Scatterplot
scatterplot_prices = sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_custmrs)
```

### Histogram

```python
# Create Histogram
hist_prices = ords_prods_custmrs['prices'].plot.hist(bins = 25)
```

##### Create sample for line chart

```python
np.random.seed(4)
```

```python
# Create a list holding True/False values to the test np.random.rant() <= 0.7
dev = np.random.rand(len(ords_prods_custmrs)) <= 0.7
```

```python
# store 70% of sample in the dataframe big
big = ords_prods_custmrs[dev]
```

```python
# store 30% of sample in the dataframe small
small = ords_prods_custmrs[~dev]
```

```python
len(ords_prods_custmrs)
```

```python
# Check sizes, if big and small adds up to main dataframe numbers
len(big) + len(small)
```

```python
# Create df for line chart, with 30% of total sample and 2 relevant columns
df_for_line = small[['orders_day_of_week','prices']]
```

### Line Chart

```python
# Create Line chart
line_orders_dow_prices = sns.lineplot(data = df_for_line, x = 'orders_day_of_week',y = 'prices')
```

# 03 Histogram Q3

```python
# Create Histogram from 'order_hour_of_day'
hist_order_hod = ords_prods_custmrs['order_hour_of_day'].plot.hist(bins = 24)
```

Most orders come in between 10am and 5pm. Midnight to 6am is a down time.

# 04 Bar Chart Q4

```python
# Create Bar Chart from 'loyalty_flag'
bar_loyalty = ords_prods_custmrs['loyalty_flag'].value_counts().plot.bar(color =['yellow', 'red', 'green'])
```

While each individual Loyal customer placed most orders (>40 each), the largest number of overall orders was placed by regular customers (roughly 16M out of a total of 32.4M), followed by Loyal and only then - New.

# 07 Scatterplot Q7

```python
# Create Scatterplot showing connection between age and earning power (income)
scatterplot_age_income = sns.scatterplot(x = 'age', y = 'income',data = ords_prods_custmrs)
```

Most people make under  200,000 (accross age groups). There is a significant increase inearning power at around 40. No one under 40 makes over 400,000, and at the same time many more people make between  200,000 and 300,000 once they reach 40+.
